```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

temperatures = np.array([22, 25, 28, 27, 30, 26, 24, 23, 25, 22, 21, 20])

mean_temp = np.mean(temperatures)
print("Mean: ", mean_temp)
plt.hist(temperatures, bins=5, edgecolor='red')
plt.title('Temperature Distribution')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()


hours_studied = np.array([1,2,3,4,5,6,7]).reshape(-1,1)
exam_scores = np.array([45,50,55,60,65,70,75])

model = LinearRegression()
model.fit(hours_studied, exam_scores)

# Predict scores
predicted_scores = model.predict(hours_studied)

# Plotting the results
plt.scatter(hours_studied, exam_scores, color='blue', label='Actual Scores')
plt.plot(hours_studied, predicted_scores, color='red', label='Predicted Scores')
plt.title('Hours Studied vs Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({
    'col_1': np.random.rand(10),
    'col_2': np.random.rand(10),
    'col_3': np.random.rand(10),
    'col_4': np.random.rand(10)
})

# First scatter plot
ax = df.plot.scatter(x="col_1", y="col_3", color="red", marker="*", s=100, label='col_1 vs col_3')

# Second scatter plot on the same axes
df.plot.scatter(x="col_2", y="col_4", color="orange", s=100, ax=ax, label='col_2 vs col_4')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Combined Scatter Plot')
plt.legend()

# Show plot
plt.show()
```
